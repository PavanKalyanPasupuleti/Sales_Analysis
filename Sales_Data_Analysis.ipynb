```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales_data={
    'Month':['Jan','Jan','Feb','Feb','Mar','Mar','Apr','Apr','May','May','Jun','Jun'],
    'Product':['Laptop','Mouse','Laptop','Keyboard','Laptop','Mouse','Keyboard','Laptop','Mouse', 'Laptop', 'Keyboard', 'Laptop'],
    'Sales':[50000,5000, 60000, 7000,55000, 4000, 8000, 65000,6000, 70000, 9000, 75000]

}

data=pd.DataFrame(sales_data)
print(data)
```

```python
total_sales=data['Sales'].sum()
print('Total Sales:',total_sales)
```

```python
monthly_sales=data.groupby('Month')['Sales'].sum()
print(monthly_sales)
```

```python
plt.figure(figsize=(8,5))
plt.plot(monthly_sales.index,monthly_sales.values,marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
product_sales=data.groupby('Product')['Sales'].sum()
print(product_sales)
```

```python
plt.figure(figsize=(8,5))
plt.bar(product_sales.index,product_sales.values)
plt.title("Product Sales")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```
